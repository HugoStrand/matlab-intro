## Koch-kurvan och snöflingan

<https://en.wikipedia.org/wiki/Koch_snowflake>

är en själv-similär fraktal.

<https://en.wikipedia.org/wiki/Self-similarity>

Den kan konstrueras iterativt genom att

1. Givet en linje mellan två punkter
2. Dela upp linjen i fyra nya linjer med 1/3 av linjens längd, enl. figur nedan.
3. För varje nytt linje-segment repetera proceduren från punkt 1.

### Uppgift 1a

Konstruera en liksidig triangel med sidlängd 2 och centrum i origo.

Beräkna x,y koordinaterna för triangelns tre hörn och samla dom i en två-dimensionell array.

Modifiera exemplet nedan som ritar en kvadrat med sidan 2 och centrum i origo

````matlab
v1 = [-1, -1];
v2 = [-1, +1];
v3 = [+1, +1];
v4 = [+1, -1];

v_square = [v1; v2; v3; v4; v1]';

x = v_square(1, :);
y = v_square(2, :);

plot(x, y, '-k')

xlim([-2, 2]); ylim([-2, 2]); axis('square');
````

### Uppgift 1b

Skriv en funktion `koch_step` som tar en kurva `v_start` som en tvådimensionell array enl. ovan och applicerar Koch:s algoritm för varje linje-segment och returnerar den nya kurvan `v`.

```octave
function v = koch_step(v_start)
    ...
end
```

Se till att din funktion ger följande resultat

````matlab
v_start = [[-1, 0] ; [+1, 0]]';

v = koch_step(v_start);
x = v(1, :); y = v(2, :);

plot(x, y, '-k')
xlim([-2, 2]); ylim([-2, 2]); axis('square');
````

````matlab
v_start = [[-1, -1] ; [+1, +1]]';

v = koch_step(v_start);
x = v(1, :); y = v(2, :);

plot(x, y, '-k')
xlim([-2, 2]); ylim([-2, 2]); axis('square');
````

### Uppgift 1c

Skriv en funktion `koch` som tar `steps` steg av iterationen med `koch_step` och returnerar den generade kurvan.

```octave
function v = koch(v_start, steps)
    ...|
end
```

Testa att din funktion ger resultatet

````matlab
steps = 6;
v_start = [[-1, -1] ; [+1, +1]]';

v = koch(v_start, steps);
x = v(1, :); y = v(2, :);

plot(x, y, '-k')
xlim([-2, 2]); ylim([-2, 2]); axis('square');
````
